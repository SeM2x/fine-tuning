```python
# 2. Load Model + Your Adapters from Drive
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "/content/drive/MyDrive/Hackathon_Qwen-Darja_Adapters", # Load FROM Drive
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)
```

```python
# 3. Convert to GGUF LOCALLY (on Colab's disk, not Drive)
# This avoids the storage limit because Colab has ~70GB free space
print("Converting locally on Colab VM...")
model.save_pretrained_gguf(
    "colab_local_gguf", # Saving to local folder
    tokenizer,
    quantization_method = "q4_k_m"
)
```

```python
# 4. Copy ONLY the final GGUF file to your Drive
# The final file is only ~5GB, so it will fit in your 11GB space!
import shutil
import os

source_file = "/content/Qwen2.5-7B-Instruct.Q4_K_M.gguf"
destination_folder = "/content/drive/MyDrive/Hackathon_Qwen-Darja_GGUF"

# Create folder if it doesn't exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

print(f"Moving {source_file} to Google Drive...")
shutil.copy(source_file, destination_folder)

print("âœ… SUCCESS! The file is now in your Google Drive folder 'Hackathon_Final_Model'")
```
